5.2 주제별 부실현황

* 피벗테이블 생성을 위한 사용자 정의 함수

```python
# 부실현황 파악을 위한 피벗데이블 생성 함수
def busil_rate(data, var):
    
    pivot = data.pivot_table(index=var, columns='부실여부', values='CNT', aggfunc=sum) # 피벗테이블 생성 
    pivot = pivot.fillna(0) # 결측값을 0으로 변경
    pivot['전체'] = pivot[1] + pivot[0]
    pivot['부실률'] = (pivot[1] / (pivot[0] + pivot[1]))*100  
    pivot.rename(columns={0:'정상',1:'부실'},inplace=True) # 변수 이름 변경
    pivot[['정상','부실','전체']] = pivot[['정상','부실','전체']].astype(int) # 표현형식을 정수형으로 변경
    pivot['부실률'] = pivot['부실률'].round(2) # 소수점 2자리에서 반올림
    globals()['pivot_{}'.format(var)] = pivot # 동적 변수 호출 (변수명을 동적으로 변경 : 어려운 문법)
    return 
```

* 체크리스트를 정리하는 함수 생성

```python
# 체크리스트 병합 초기 테이블 생성
check_table = base_data2.loc[:,['체크리스트저촉1','부실여부','CNT']]
check_table.rename(columns={'체크리스트저촉1':'체크리스트저촉'}, inplace=True)
del_idx = check_table[check_table['체크리스트저촉']==' '].index
check_table = check_table.drop(del_idx)
check_table = check_table.dropna()
check_table.info()
```

```python
# 체크리스트 병합 함수
def check_concat(table, var):
    global check_table
    # 칼럼 선택
    check_table_tmp = base_data2.loc[:,[var,'부실여부','CNT']]
    # 변수이름 변경 : 통일화
    check_table_tmp.rename(columns={var:'체크리스트저촉'}, inplace=True)
    # 삭제대상 인덱스 정의 : 공백만 있는 row 삭제
    del_idx = check_table_tmp[check_table_tmp['체크리스트저촉']==' '].index
    # 행삭제
    check_table_tmp = check_table_tmp.drop(del_idx)
    # 결측값 삭제
    check_table_tmp = check_table_tmp.dropna()
    # 데이터 병합
    check_table = pd.concat([table, check_table_tmp])
    print(check_table.info())
    return check_table
```
